## Numpy

```python
import numpy as np
```

### Numpy Arrays
* How to declare an array
* How to access array elements using indexing

```python
a = np.array([1,2,3,4,5])
print(a[0]) 
print(a[4])
type(a)
```

### Properties of Array
* __Size__ Returns the size (number of elements) of an array
* __ndim__ Returns the dimensions (1D, 2D etc) of an array
* __shape__ Returns the shape (rows, cols) of an array

```python
print(a)
print(a.size)
print(a.ndim)
print(a.shape)
```

### Two Dim Arrays and it properties

```python
b = np.array([ [1,2],
               [3,4],
               [5,6] ])

print(b)
print(b.size)
print(b.ndim)
print(b.shape)
```

### Reshaping the array
* __reshape()__ function can be used to reshape/redfine the number of rows and cols

```python
c = np.array([ 1,12,3,44,5,6,67,8])

d = c.reshape(4,2)

print("Array c : ", c)
print("Array d : ", d)
print("c.shape : ", c.shape)
print("d.shape : ", d.shape)
```

### Array elements can be modified

```python
c = np.array([ 1,2,3,4,5,6,7,8])

print("Before change in Array c : ", c)
c[0] = 100
print("After change in Array c : ", c)
```

### Slicing the array

```python
c = np.array([1,2,3,4,5,6,7,8])

d =  c[1:4]

print(d)
```

```python
c[3:5] = 40,50

print(c)

print(d)
```

### Performing operations on arrays 
* Addition using Numpy Arrays

```python
A = np.array([1,0])
B = np.array([0,1])
C = A + B
C
```

### Subtraction & Scaler Multiplication using Numpy

```python
A = np.array([1,0])
B = np.array([0,1])
D = A - B
D
```

```python
A = np.array([2,3])
E = 2*A
E
```

### Hadamard Product (Element-wise Multiplication)

```python
A = np.array([2,3])
B = np.array([4,5])
F = A* B
F
```

### Matrix Multiplication

```python
A = np.array([[2,3]])
B = np.array([[4],[5]])
G = np.dot(A,B)
G
```

### Broadcasting

```python
A = np.array([2,6,8,7,3])
H = A + 4
H
```

### Universal Functions
* Various functions can be applied on numpy arrays

```python
A = np.array([2,6,8,7,3])
meanA = A.mean()
maxA = A.max()

print(meanA)
print(maxA)
print(np.pi)
```

```python
J = np.array([0, np.pi/2, np.pi])
J
```

```python
K = np.sin(J)
K
```

### linspace Function
* __linspace()__ Returns number spaces evenly w.r.t interval.
* The following code returns a 5 element numpy array in the range (-2, 2)

```python
np.linspace(2,8,num=5) # 2,8 means that in the range of 2 to 8 we need 5 numbers with equal distance 
```

* The following code returns a 9 element numpy array in the range (-2, 2)

```python
np.linspace(2,8,num=9) # 2,8 means that in the range of 2 to 8 we need 9 numbers with equal distance 
```

### Ploting Charts in Python

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,2*np.pi,100) #It means that from 0 to 2pi we need values 100 values with equal distance 
y = np.sin(x)                  #In x we have the the array with the 100 values of 0 to 2pi...this y will find the sin value of each value

plt.plot(x,y)   # This will plot the graph of x and y values...
```

```python
a = np.random.random(100)
b = np.random.random(100)

print(a)
```

```python
print(b)
```

```python
plt.scatter(a,b)
plt.show()
```

```python
plt.hist(a,bins=30)
plt.show()
```

```python
plt.hist(b,bins=30)
plt.show()
```

```python

# Generate data for the first line
x1 = np.linspace(0, 10, 20) # From zero to 10 we will have 20 equally distanced values 
y1 = np.sin(x1)          # sin values of each "x1 numpy array"

# Generate data for the second line
x2 = np.linspace(0, 10, 20)
y2 = np.cos(x2)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the first line with a square marker
ax.plot(x1, y1, '-s', label='sin')

# Plot the second line with a diamond marker
ax.plot(x2, y2, '-D', label='cos')



# Add a legend to the plot
ax.legend()

# Set the title and labels
ax.set_title("Sine and Cosine")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Show the plot
plt.show()
```

## NetworkX

```python
import networkx as nx
```

```python
# Create an empty graph
G = nx.Graph()
```

```python
# Adding nodes manually
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
```

```python
G.add_nodes_from([5,6])
```

```python
# Adding an edge
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(4,5)
G.add_edge(5,6)
```

```python
# Drawing graph
nx.draw(G)
```

```python
# Adding edge between non-existent nodes
G.add_edge(1,9)
```

```python
# Adding edges from a list
G.add_edges_from ([(1,4), (2,5)])
```

```python
# Drawing graph
nx.draw(G)
```

```python
# Drawing options
options = {
    "font_size": 10,
    "node_size": 500,
    "node_color": "white",
    "edgecolors": "blue",
    "linewidths":4,
    "width": 4,
}
```

```python
# Drawing with options
nx.draw_networkx(G, **options)
```

```python
# Writing graph to a text file
nx.write_edgelist(G,"graph.txt")
```

```python
# Print number of nodes
print(G.number_of_nodes())
```

```python
# Print number of edges
print(G.number_of_edges())
```

```python
# Draw graph with labels
nx.draw(G, with_labels=True)
```

```python
# Degrees of node 1 and 3
G.degree([1,3]) 
```
