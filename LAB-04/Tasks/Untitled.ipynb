## Breadth first serach  

```python
graph = {
  '5' : ['3','7'],
  '3' : ['2', '4'],
  '7' : ['8'],
  '2' : [],
  '4' : ['8'],
  '8' : []
}

visited = [] # List for visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node): #function for BFS
    
    visited.append(node)
    queue.append(node)

   
    while queue:          # Creating loop to visit each node
        m = queue.pop(0) 
        print (m, end = " ") 

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("Following is the Breadth-First Search")
bfs(visited, graph, '5')
```

## Binary Search Tree using DFS 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def dfs(node):
        if node is None:
            return
        
        print(node.data)

        dfs(node.left)
        dfs(node.right)

# Create a binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Perform DFS traversal
dfs(root)
```

## UNIFORM COST SEARCH 

```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Initialize the priority queue with the start node and its cost
    priority_queue = [(0, start)]
    # Initialize the visited set to track visited nodes
    visited = set()
    # Initialize the parent dictionary to track the path
    parent = {start: None}
    # Initialize the cost dictionary to track the cost of reaching each node
    cost = {start: 0}

    while priority_queue:
        # Pop the node with the minimum cost from the priority queue
        current_cost, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            # Reconstruct and return the path and the minimum cost
            path = construct_path(parent, goal)
            return path, cost[goal]

        if current_node in visited:
            # Skip the node if it has already been visited
            continue

        visited.add(current_node)

        for neighbor, edge_cost in graph[current_node].items():
            # Calculate the new cost to reach the neighbor node
            new_cost = cost[current_node] + edge_cost

            if neighbor not in cost or new_cost < cost[neighbor]:
                # Update the cost and parent for the neighbor node
                cost[neighbor] = new_cost
                parent[neighbor] = current_node
                # Add the neighbor node to the priority queue with its cost
                heapq.heappush(priority_queue, (new_cost, neighbor))

    # No path found from the start to the goal
    return [], float('inf')

def construct_path(parent, goal):
    path = [goal]
    current = goal

    while parent[current]:
        current = parent[current]
        path.append(current)

    path.reverse()
    return path
```

```python
# Example usage
graph = {
    'A': {'B': 2, 'C': 5},
    'B': {'D': 3},
    'C': {'D': 1, 'E': 6},
    'D': {'E': 2},
    'E': {}
}

start = 'A'
goal = 'E'

path, min_cost = uniform_cost_search(graph, start, goal)
print("Path:", path)
print("Minimum cost:", min_cost)
```

```python
def least_cost_path(graph, start, goal, m):
    # Initialize the priority queue with the start node and its cost
    priority_queue = [(0, start)]
    # Initialize the visited set to track visited nodes
    visited = set()
    # Initialize the parent dictionary to track the path
    parent = {start: None}
    # Initialize the cost dictionary to track the cost of reaching each node
    cost = {start: 0}

    while priority_queue:
        # Pop the node with the minimum cost from the priority queue
        current_cost, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            # Reconstruct and return the path and the minimum cost
            path = construct_path(parent, goal)

            # Check if the path has exactly m edges
            if len(path) == m:
                return path, cost[goal]

        if current_node in visited:
            # Skip the node if it has already been visited
            continue

        visited.add(current_node)

        for neighbor, edge_cost in graph[current_node].items():
            # Calculate the new cost to reach the neighbor node
            new_cost = cost[current_node] + edge_cost

            if neighbor not in cost or new_cost < cost[neighbor]:
                # Update the cost and parent for the neighbor node
                cost[neighbor] = new_cost
                parent[neighbor] = current_node
                # Add the neighbor node to the priority queue with its cost
                heapq.heappush(priority_queue, (new_cost, neighbor))

    # No path found from the start to the goal
    return [], float('inf')

def construct_path(parent, goal):
    path = [goal]
    current = goal

    while parent[current]:
        current = parent[current]
        path.append(current)

    path.reverse()
    return path


# Example usage
graph = {
    '0': {'1': 5, '6': -1},
    '1': {'2': 7, '5': 5, '6': 3},
    '2': {'8': 3},
    '3': {'4': 10},
    '4': {},
    '5': {'4': 1, '2': -1, '3': 9},
    '6': {'5': 0},
    '7': {},
    '8': {}
}

start = '0'
goal = '3'
m = 4

path, min_cost = least_cost_path(graph, start, goal, m)
print("Path:", path)
print("Minimum cost:", min_cost)
```

```python

from collections import deque
initial_state = [[1, 2, 3], [4, 0, 5], [6, 7, 8]]
goal_state = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]

def get_moves(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                if i > 0:
                    new_state = [row[:] for row in state]
                    new_state[i][j], new_state[i-1][j] = new_state[i-1][j], new_state[i][j]
                    yield new_state
                if i < 2:
                    new_state = [row[:] for row in state]
                    new_state[i][j], new_state[i+1][j] = new_state[i+1][j], new_state[i][j]
                    yield new_state
                if j > 0:
                    new_state = [row[:] for row in state]
                    new_state[i][j], new_state[i][j-1] = new_state[i][j-1], new_state[i][j]
                    yield new_state
                if j < 2:
                    new_state = [row[:] for row in state]
                    new_state[i][j], new_state[i][j+1] = new_state[i][j+1], new_state[i][j]
                    yield new_state

def bfs(initial_state, goal_state):
    visited_states = set()
    queue = deque([(initial_state, [])])
    while queue:
        state, path = queue.popleft()
        if state == goal_state:
            return path
        for move in get_moves(state):
            if tuple(map(tuple, move)) not in visited_states:
                visited_states.add(tuple(map(tuple, move)))
                queue.append((move, path + [move]))

path = bfs(initial_state, goal_state)
for i in path:
    for j in i:
        for k in j:
            print(k,end=" ")
        print()
    
    print()

```
